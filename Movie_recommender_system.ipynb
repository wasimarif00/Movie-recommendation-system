```python
import numpy as np
import pandas as pd
```

```python
movies=pd.read_csv("/content/tmdb_5000_movies.csv")
```

```python
credits=pd.read_csv("/content/tmdb_5000_credits.csv")
```

```python
movies.head()
```

```python
movies.shape
```

```python
credits.head()
```

```python
credits.shape
```

```python
#Merged the two DataFrames, `movies` and `credits`, based on the common column "title"
movies = movies.merge(credits,on='title')
```

```python
movies.head()
```

```python
#Select only the relevant columns from the merged DataFrame

movies = movies[['movie_id','title','overview','genres','keywords','cast','crew']]
```

```python
movies.head()
```

```python
movies.dropna(inplace=True)
```

```python
# Convert the `genres` column from a string of dictionaries to a list of genre names
import ast

movies['genres'] = movies['genres'].apply(lambda x: [i['name'] for i in ast.literal_eval(x)])
```

```python
movies['keywords'] = movies['keywords'].apply(lambda x: [i['name'] for i in ast.literal_eval(x)])
```

```python
movies.head()
```

```python
movies['cast'] = movies['cast'].apply(lambda x: [i['character'] for i in ast.literal_eval(x)])
```

```python
movies.head()
```

```python
#This modification is done for each element of the `cast` column, so that the final DataFrame contains only the names of the first three cast members for each movie

movies['cast'] = movies['cast'].apply(lambda x:x[0:3])
```

```python
movies.head()
```

```python
# From crew fetch the director name

movies['crew'] = movies['crew'].apply(lambda x: [i['name'] for i in ast.literal_eval(x) if i['job'] == 'Director'])
movies.head()
```

```python
#Function removes spaces from each element of a list

def collapse(L):
  L1 = []
  for i in L:
    L1.append(i.replace(" ", ""))
  return L1
```

```python
movies['cast'] = movies['cast'].apply(collapse)
movies['crew'] = movies['crew'].apply(collapse)
movies['genres'] = movies['genres'].apply(collapse)
movies['keywords'] = movies['keywords'].apply(collapse)
```

```python
movies.head()
```

```python
movies['overview'] = movies['overview'].apply(lambda x:x.split())
```

```python
movies.head()
```

```python
# Creates a new column called "tags" that contains a comprehensive list of words related to each movie, including the overview, genres, keywords, cast, and crew

movies['tags'] = movies['overview'] + movies['genres'] + movies['keywords'] + movies['cast'] + movies['crew']
```

```python
new_df = movies.drop(columns=['overview','genres','keywords','cast','crew'])
```

```python
new_df.head()
```

```python
new_df['tags'] = new_df['tags'].apply(lambda x: " ".join(x))
```

```python
new_df.head()
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features=5000,stop_words='english')
```

```python
vector = cv.fit_transform(new_df['tags']).toarray()
```

```python
from sklearn.metrics.pairwise import cosine_similarity
similarity = cosine_similarity(vector)
```

```python
def recommend(movie):
    index = new_df[new_df['title'] == movie].index[0]
    distances = sorted(list(enumerate(similarity[index])),reverse=True,key = lambda x: x[1])
    for i in distances[1:6]:
        print(new_df.iloc[i[0]].title)
```

```python
recommend('John Carter')
```
